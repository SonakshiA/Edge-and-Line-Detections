**EDGE DETECTIONS**

1. Vertical Edge Detection using Sobel Operator

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('tetris.jpeg', cv2.IMREAD_GRAYSCALE)

# Apply the Sobel filter for vertical edge detection
sobel_vertical = cv2.Sobel(image, cv2.CV_64F, dx=1, dy=0, ksize=3)

# Convert the result to an 8-bit image
sobel_vertical = cv2.convertScaleAbs(sobel_vertical)

# Display the original and edge-detected images
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_vertical, cmap='gray')
plt.title('Vertical Edge Detection')
plt.axis('off')

plt.show()
```

**LINE DETECTIONS**

Kernel for detecting lines at +45 degrees

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('House.jpeg', cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
edges = cv2.Canny(image, 50, 150)

# Create a kernel to detect lines at +45 degrees
kernel = np.array([[-1, -1, 2],
                   [-1, 2, -1],
                   [2, -1, -1]])

# Apply the filter to the edge-detected image
line_detection = cv2.filter2D(edges, -1, kernel)

# Convert the result to an 8-bit image for display
line_detection_abs = cv2.convertScaleAbs(line_detection)

# Thresholding to get binary image
_, binary_image = cv2.threshold(line_detection_abs, 50, 255, cv2.THRESH_BINARY)

# Display the images
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Line Detection at +45 Degrees')
plt.axis('off')

plt.show()
```
